```python
img1=cv2.imread("./test3.jpg") #logo
img2=cv2.imread("./test4.jpg") #logo
img3=cv2.imread("./test2.jpg") 
rows,cols=img1.shape[:2] 
range=img2[:rows,:cols]  #ROIlogo
img1_gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
print(np.array(img1_gray)[144])
#cv2.threshold()cv2.cvtColor()
ret,mask=cv2.threshold(img1_gray,175,255,cv2.THRESH_BINARY)
#
mask_inv=cv2.bitwise_not(mask) #mask
roi1=cv2.bitwise_and(img1,img1,mask_inv) 
#  #ROI1img2logo
roi2=cv2.bitwise_and(img2,img2,mask)
# ROI2logologo
add=cv2.add(roi1,roi2)  # ROI
img2[:rows,:cols]=add  # img2
# #
cv2.imshow("mask",mask)
# cv2.imshow("mask",mask)
cv2.imshow("mask_inv",mask_inv)
cv2.imshow("add1",roi1)
# # cv2.imshow("add2",add2)
# cv2.imshow("add",add)
# cv2.imshow("img2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
