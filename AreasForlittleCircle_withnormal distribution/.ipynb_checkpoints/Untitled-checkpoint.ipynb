```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import math
from PIL import Image
from tqdm import tqdm
import cv2
```

```python
i = 500
cov = 0
N=0
n=0
R = 0.5
r = 0.06
img2=cv2.imread("./test3.jpg") #logo图像
rows,cols,channels=img2.shape
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret,mask = cv2.threshold(img2gray,175,255,cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
c = Circle(xy=(0,0),radius=0.5,color = '#FF0000',edgecolor='None')
ax.add_patch(c)
plt.axis("equal")
plt.axis("off")
list_kn = []
for p in tqdm(range(0,500)):
    mean = np.array([0,0])
    conv = np.array([[0.259,0.0],[0,0.259]])
    op = 0.98
    s = 0
    redl = 0
    blackl = 0
    while s<=0.98:
        x,y = np.random.multivariate_normal(mean=mean,cov=conv,size=1).T
        z = [x[0],y[0]]
        c2 = Circle(xy=z,radius=0.06,color = '#000000',edgecolor='None')
        N = N + 1
        ax.add_patch(c2)
        d = math.sqrt(z[0]*z[0]+z[1]*z[1])
        if (d <= 0.5-0.06) or (d>0.5-0.06 and d<0.5+0.06):
#             c1 = Circle(xy=z,radius=0.06,color = '#000000',edgecolor='None')
#             ax.add_patch(c1)
            n = n + 1
            plt.savefig('./test2.jpg')
            img1=cv2.imread("./test2.jpg")
            roi = img1[0:rows,0:cols]
            img1_bg = cv2.bitwise_and(roi,roi,mask=mask_inv)
            img2_fg = cv2.bitwise_and(img2,img2,mask = mask)
            dst = cv2.add(img1_bg,img2_fg)
            img1[0:rows,0:cols] = dst
            plt.imshow(img1[...,::-1])
            imfg = Image.fromarray(img1[...,::-1])
#             imfg.save('./tesr.jpg')
#             print(imfg)
#             plt.savefig('./tesr.jpg')
#             cvb = Image.open('./tesr.jpg')
#             plt.savefig('./tesr.jpg')
#             cvb = Image.open('./tesr.jpg')
#             print(cvb1)
            for i in tqdm(range(imfg.size[0])):
                for j in range(imfg.size[1]):
                    k = imfg.getpixel((i,j))
                    if k[0]+k[1]+k[2]!=3*55:
                        if k[0]+k[1]+k[2] >= 0 and k[0]+k[1]+k[2]<=60:
                            print(k[0]+k[1]+k[2])
                            blackl += 1
                        elif k[0]+k[1]+k[2] >= 250 and k[0]+k[1]+k[2]<=320:
                            print(k[0]+k[1]+k[2])
                            redl += 1
            print(blackl,redl)   
            s = blackl/(blackl+redl)
            print(f'小圆个数为{n},面积为{s}')
    print(f'第{p}次循环执行完成')
    break
    list_kn.append((N,n))
    N=0
    n=0
print(list_kn)
```
